````python
# Core Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# Deep Learning Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import EfficientNetB0

# Evaluation Metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Enable mixed precision for faster training on GPUs
tf.keras.mixed_precision.set_global_policy('mixed_float16')

print(f"TensorFlow Version: {tf.__version__}")
print(f"GPU Available: {len(tf.config.list_physical_devices('GPU')) > 0}")
print(f"Mixed Precision: Enabled (for faster training)")
````

````python
# IMPORTANT: Update this path to your dataset location
BASE_DIR = r'C:\Users\utkar\Desktop\labmentix project\birth and drone detection\classification_dataset'
TRAIN_DIR = os.path.join(BASE_DIR, 'train')
VALID_DIR = os.path.join(BASE_DIR, 'valid')
TEST_DIR = os.path.join(BASE_DIR, 'test')

# Optimized Configuration for Fast Training
IMG_SIZE = 224
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
BATCH_SIZE = 32  # Increase to 64 if you have good GPU
EPOCHS = 30  # Reduced epochs with early stopping for faster training
LEARNING_RATE = 0.001
FINE_TUNE_LR = LEARNING_RATE / 10  # Lower LR for fine-tuning
CLASS_NAMES = ['bird', 'drone']
NUM_CLASSES = len(CLASS_NAMES)

print(f"\n‚öôÔ∏è Configuration:")
print(f"  Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Epochs: {EPOCHS}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Fine-tune LR: {FINE_TUNE_LR}")
print(f"  Classes: {CLASS_NAMES}")
````

````python
def count_images(directory):
    """Count images in each class folder"""
    counts = {}
    for class_name in CLASS_NAMES:
        class_path = os.path.join(directory, class_name)
        if os.path.exists(class_path):
            counts[class_name] = len([f for f in os.listdir(class_path)
                                      if f.endswith(('.jpg', '.jpeg', '.png'))])
        else:
            counts[class_name] = 0
    return counts

train_counts = count_images(TRAIN_DIR)
valid_counts = count_images(VALID_DIR)
test_counts = count_images(TEST_DIR)

# Create summary DataFrame
data_summary = pd.DataFrame({
    'Train': train_counts,
    'Validation': valid_counts,
    'Test': test_counts
})

print("\nüìä Dataset Summary:")
print(data_summary)
print(f"\nTotal Images: {sum(train_counts.values()) + sum(valid_counts.values()) + sum(test_counts.values())}")

# Visualize distribution
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, (title, counts) in enumerate([('Training', train_counts), ('Validation', valid_counts), ('Test', test_counts)]):
    axes[idx].bar(counts.keys(), counts.values(), color=['#3498db', '#e74c3c'])
    axes[idx].set_title(f'{title} Set', fontsize=14, fontweight='bold')
    axes[idx].set_ylabel('Number of Images')
    for i, v in enumerate(counts.values()):
        axes[idx].text(i, v + max(counts.values())*0.02, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# Training Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

# Validation and Test - Only Rescaling
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create generators with explicit RGB color mode
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=42,
    color_mode='rgb'  # Ensure 3-channel RGB images
)

valid_generator = valid_datagen.flow_from_directory(
    VALID_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    color_mode='rgb'
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    color_mode='rgb'
)

print(f"\n‚úÖ Data Generators Created!")
print(f"Training: {train_generator.samples} samples")
print(f"Validation: {valid_generator.samples} samples")
print(f"Test: {test_generator.samples} samples")
print(f"Class Indices: {train_generator.class_indices}")
````

````python
# Manual cache clearing (run this first if error persists)
import os
import shutil

# Clear Keras cache
cache_dir = os.path.expanduser('~/.keras/models')
print(f"üìÅ Keras cache directory: {cache_dir}")

if os.path.exists(cache_dir):
    efficientnet_files = [f for f in os.listdir(cache_dir) if 'efficientnet' in f.lower()]
    if efficientnet_files:
        print(f"üóëÔ∏è Found {len(efficientnet_files)} EfficientNet cache files")
        for f in efficientnet_files:
            file_path = os.path.join(cache_dir, f)
            try:
                os.remove(file_path)
                print(f"   ‚úÖ Removed: {f}")
            except Exception as e:
                print(f"   ‚ùå Could not remove {f}: {e}")
        print("\n‚úÖ Cache cleared! Now run the model building cell.")
    else:
        print("‚ÑπÔ∏è No EfficientNet cache files found")
else:
    print("‚ÑπÔ∏è Cache directory doesn't exist yet")
from tensorflow.keras.applications import MobileNetV2

def build_transfer_model():
    """Build Transfer Learning model with MobileNetV2"""
    tf.keras.backend.clear_session()
    
    # Load MobileNetV2 base model (alternative to EfficientNetB0)
    base_model = MobileNetV2(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )
    
    base_model.trainable = False
    
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(NUM_CLASSES, activation='softmax', dtype='float32')
    ], name='Transfer_MobileNetV2')
    
    return model, base_model

# Build model
print("\nüèóÔ∏è Building Transfer Learning Model (MobileNetV2)...")
transfer_model, base_model = build_transfer_model()

transfer_model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n‚úÖ Model Built Successfully!")
transfer_model.summary()
````

````python
os.makedirs('models', exist_ok=True)

# Callbacks for initial training
initial_callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ModelCheckpoint('models/transfer_initial.h5', monitor='val_accuracy',
                    save_best_only=True, mode='max', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
]

print("\nüöÄ Phase 1: Initial Training (Base Frozen)...\n")

history_initial = transfer_model.fit(
    train_generator,
    epochs=15,  # Initial training with frozen base
    validation_data=valid_generator,
    callbacks=initial_callbacks,
    verbose=1
)

print("\n‚úÖ Initial Training Complete!")
````

````python
# Unfreeze last 30 layers for fine-tuning
base_model.trainable = True

# Freeze all except last 30 layers
for layer in base_model.layers[:-30]:
    layer.trainable = False

print(f"\nüîì Unfrozen last 30 layers for fine-tuning")
print(f"Trainable layers: {sum([1 for layer in transfer_model.layers if layer.trainable])}")

# Recompile with lower learning rate
transfer_model.compile(
    optimizer=Adam(learning_rate=FINE_TUNE_LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks for fine-tuning
finetune_callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1),
    ModelCheckpoint('models/transfer_finetuned.h5', monitor='val_accuracy',
                    save_best_only=True, mode='max', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-8, verbose=1)
]

print("\nüöÄ Phase 2: Fine-Tuning (Last 30 Layers Unfrozen)...\n")

history_finetune = transfer_model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=valid_generator,
    callbacks=finetune_callbacks,
    verbose=1
)

print("\n‚úÖ Fine-Tuning Complete!")
````

````python
# Plot training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Accuracy plot
axes[0].plot(history_initial.history['accuracy'], label='Initial Train', marker='o', color='blue')
axes[0].plot(history_initial.history['val_accuracy'], label='Initial Val', marker='s', color='cyan')
axes[0].plot(range(len(history_initial.history['accuracy']), 
                  len(history_initial.history['accuracy']) + len(history_finetune.history['accuracy'])),
             history_finetune.history['accuracy'], label='Finetune Train', marker='o', color='green')
axes[0].plot(range(len(history_initial.history['val_accuracy']), 
                  len(history_initial.history['val_accuracy']) + len(history_finetune.history['val_accuracy'])),
             history_finetune.history['val_accuracy'], label='Finetune Val', marker='s', color='lime')
axes[0].axvline(x=len(history_initial.history['accuracy']), color='red', linestyle='--', label='Fine-tune start')
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Loss plot
axes[1].plot(history_initial.history['loss'], label='Initial Train', marker='o', color='blue')
axes[1].plot(history_initial.history['val_loss'], label='Initial Val', marker='s', color='cyan')
axes[1].plot(range(len(history_initial.history['loss']), 
                  len(history_initial.history['loss']) + len(history_finetune.history['loss'])),
             history_finetune.history['loss'], label='Finetune Train', marker='o', color='green')
axes[1].plot(range(len(history_initial.history['val_loss']), 
                  len(history_initial.history['val_loss']) + len(history_finetune.history['val_loss'])),
             history_finetune.history['val_loss'], label='Finetune Val', marker='s', color='lime')
axes[1].axvline(x=len(history_initial.history['loss']), color='red', linestyle='--', label='Fine-tune start')
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

best_val_acc = max(history_finetune.history['val_accuracy'])
best_epoch = np.argmax(history_finetune.history['val_accuracy']) + 1
print(f"\nüìä Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%) at epoch {best_epoch}")
````

````python
# Load best model
best_model = load_model(r'C:\Users\utkar\Desktop\labmentix project\birth and drone detection\models/transfer_finetuned.h5')

print("\nüìä Evaluating Model on Test Set...\n")
test_loss, test_accuracy = best_model.evaluate(test_generator, verbose=1)

print(f"\n‚úÖ Test Results:")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
````

````python
# Get predictions
y_pred_probs = best_model.predict(test_generator, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_generator.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[c.upper() for c in CLASS_NAMES],
            yticklabels=[c.upper() for c in CLASS_NAMES],
            cbar_kws={'label': 'Count'})
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontweight='bold')
plt.xlabel('Predicted Label', fontweight='bold')
plt.tight_layout()
plt.show()

# Classification Report
print("\nüìã Classification Report:")
print(classification_report(y_true, y_pred, target_names=[c.upper() for c in CLASS_NAMES]))

# Calculate metrics
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"\nüìä Summary Metrics:")
print(f"Accuracy:  {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")
````
